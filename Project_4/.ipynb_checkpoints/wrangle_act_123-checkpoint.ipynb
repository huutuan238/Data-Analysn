```python
# Create tweet-json.txt file to store obtained information
file_name = 'tweet-json.txt'
save_response_to_file(res, file_name)
```

```python
# Create a function to load special data to data frame
def load_json_content_file_to_df(file_path, columns):
    # Read downloaded txt file line by line into a pandas DataFrame
    datas = []
    with open(file_path, 'r') as f:
        lines = f.readlines()
        for line in lines:
            obj = json.loads(line)
            datas.append({col: obj.get(col) for col in columns})

    return pd.DataFrame(datas, columns=columns)
```

```python
tweet_df = load_json_content_file_to_df(file_name, ['id', 'retweet_count', 'favorite_count'])
tweet_df.rename(columns={'id': 'tweet_id'}, inplace=True)
tweet_df.head()
```

b. Only the most confident prediction for the image is required.

        Define

- I will collect confidence index by taking priority from p1_dog > p2_dog > p3_dog

        Code

```python
def collect_breed_with_priority_func(df, is_return_p_col):
    result = 'Unknown_breed', 0
    if df['p1_dog'] == True:
        result =  df['p1'], df['p1_conf']
    elif df['p2_dog'] == True:
        result =  df['p2'], df['p2_conf']
    elif df['p3_dog'] == True:
        result =  df['p3'], df['p3_conf']

    return result[0] if is_return_p_col else result[1]
     
        
# Apply the function by column        
image_prediction_df['breed'] = image_prediction_df.apply(collect_breed_with_priority_func, axis=1, is_return_p_col=True)
image_prediction_df['confident_ability'] = image_prediction_df.apply(collect_breed_with_priority_func, axis=1, is_return_p_col=False)

# Only keep useful columns
useful_cols = ['tweet_id', 'jpg_url', 'breed', 'confident_ability']
image_prediction_df = image_prediction_df[useful_cols]
```

## V.Analyzing Data & Visualization

1. The popularity of dog breeds can be determined by the number of original tweets, total retweets, and total favorites they receive.

- let find out top 10 popular dog breads base on quantity of posts

```python
top_ten_bread = twitter_archive_enhanced_df['breed'].value_counts()[10::-1]

top_ten_bread.plot(kind = 'barh', title = 'Top 10 popular dog breeds based on quantity of Posts')
plt.xlabel('Quantity of Posts')
plt.ylabel('Dog Breeds')
```

We have refined the image_prediction_df data to exclusively include tweets pertaining to dogs. Based on the graph, it is evident that the Golden Retriever holds the top spot as the most sought-after dog breed with a total about ~160 tweets. Following closely behind is the Labrador Retriever, ranking second in popularity with ~100 tweets. The subsequent three breeds in popularity are Pembroke, Chihuahua, and Pug, respectively. After these breeds, there are various other breeds listed in descending order. 

Our next step is to plot a visual representation that illustrates the popularity of different dog breeds based on the number of retweets and favorites they have received.

```python
def get_top_10_total_value_base_on_bread(df, target_column):
    df = df[['breed', 'retweet_count', 'favorite_count']]
    df_group = df.groupby(by=['breed'])
    df_target_sum = df_group[target_column].agg('sum')
    df_target_sum.sort_values(ascending=False, inplace=True)
    return df_target_sum[10::-1]
```

```python
# set to show 2 char on one plot
f, (ax1, ax2) = plt.subplots(2, 1)

top_10_breed_retweet_df = get_top_10_total_value_base_on_bread(twitter_archive_enhanced_df, 'retweet_count')
top_10_breed_retweet_df.plot.barh(ax=ax1, figsize=(7,10), color='#6A5ACD')
ax1.set_title("Top 10 Breeds Based On Retweets Count")

top_10_breed_favorite_df = get_top_10_total_value_base_on_bread(twitter_archive_enhanced_df, 'favorite_count')
top_10_breed_favorite_df.plot.barh(ax=ax2, color='#2E8B57')
ax2.set_title("Top 10 Breeds Based on Favorite Count")

f.subplots_adjust(hspace=0.5)
```

The list of the top 10 popular breeds based on tweet count includes Golden_retriever, Labrador_retriever, Pembroke, Chihuahua, Pug, Toy Poodle, Chow, Samoyed, Pomeranian, and Malamute, ranked in descending order. When it comes to retweet count, the ranking remains similar, with Golden_retriever, Labrador_retriever, Pembroke, Chihuahua, Samoyed, French_bulldog, Cocker Spaniel, Chow, Pug, Toy Poodle, and Pomeranian making up the top 10 breeds. The top 10 breeds based on favorite count follow a similar trend, except for a few variations. Specifically, the 5th breed is French_bulldog, the 6th breed is Chow, the 7th breed is Cocker_spaniel, and the 10th breed is Malamute. This resemblance between retweets and favorites can be attributed to the likelihood of people who retweet also favoriting the tweet. Overall, the consistent pattern observed across all three graphs confirms that Golden_retriever, Labrador_retriever, Pembroke, and Chihuahua dominate as the most popular dog breeds, with the Golden_retriever securing the top position as the most favored breed overall.

### 4. The tweet count determines the average ratings for popular dog breeds

```python
top_10_tweet_count_series = twitter_archive_enhanced_df['breed'].value_counts().sort_values(ascending=False).nlargest(10)
top_10_tweet_count_df = top_10_tweet_count_series.rename_axis('dogs_breed').reset_index(name='count')
top_10_tweet_count_df
```

```python
top_avg_rating_series = twitter_archive_enhanced_df.groupby('breed').rating_numerator.mean().sort_values(ascending=False)
top_avg_rating_df = top_avg_rating_series.rename_axis('breed').reset_index(name='average_rating')
top_avg_rating_df
```

```python

# Filter top avarage rating in dog breed that in top 10 tweet count 
top_breed_avg_rating = top_avg_rating_df[top_avg_rating_df['breed'].isin(top_10_tweet_count_df['dogs_breed'])]
top_breed_avg_rating
```

```python
# Show Plot chart
top_breed_avg_rating.plot(kind='bar', x='breed', y='average_rating')  
plt.ylabel("Average ating")
plt.xlabel("Dog Breeds");
plt.title("The Top 10 Dog Breeds' average ratings base on number of tweets")
```

The ratings for popular dogs exhibit a remarkable similarity among them.
